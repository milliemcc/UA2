```python
# Here is the code for challenge 2, Lab 2

import requests
import pandas as pd
import geopandas as gpd


# the following code is taken from the example and is the preprocessing for the bikes data
url_bikes = "https://api.glasgow.gov.uk/mobility/v1/get_rentals?startDate=2022-05-01&endDate=2023-05-01"
# Making the query to the web server, using the Get method from the requests library 
response = requests.get(url_bikes)
response
```

```python
data = response.json()
data
```

```python
rental_data = data['data']
rental_data
```

```python
rental_pd = pd.DataFrame(rental_data)

clean_rental_pd = rental_pd.dropna(subset=['startPlaceLat', 'startPlaceLong', 'endPlaceLat','endPlaceLong'])
```

```python
gdf_bikes_end = gpd.GeoDataFrame(clean_rental_pd, geometry=gpd.points_from_xy(clean_rental_pd['endPlaceLong'], clean_rental_pd['endPlaceLat']))
```

```python
# checking for any nan values in the coordinates 

nan_in_column_Lat = rental_pd['endPlaceLat'].isna().any()
nan_in_column_Long = rental_pd['endPlaceLong'].isna().any()

print(nan_in_column_Lat,nan_in_column_Lat)
```

```python
# remove nan values from the dataset and check to see how many have been dropped and the datatype of columns

clean_rental_pd = rental_pd.dropna(subset=['endPlaceLat','endPlaceLong'])
clean_rental_pd.info()
```

```python
# convert to a gdf using geopandas and using the coordinates to create a point geometry
gdf_bikes_end = gpd.GeoDataFrame(clean_rental_pd, geometry=gpd.points_from_xy(clean_rental_pd['endPlaceLong'], clean_rental_pd['endPlaceLat']))
```

```python
# convert to a gdf using geopandas and using the coordinates to create a point geometry
gdf_bikes_end = gpd.GeoDataFrame(clean_rental_pd, geometry=gpd.points_from_xy(clean_rental_pd['endPlaceLong'], clean_rental_pd['endPlaceLat']))
```

```python
# filter unwanted columns from the gdf
keep_cols = [
    "endDate",
    "endPlaceId",
    "endPlaceName",
    "durationSeconds",
    "isInvalid",
    "price",
    "isEbike",
    "endPlaceLat",
    "endPlaceLong",
    "geometry",
]
gdf_bikes_end = gdf_bikes_end[keep_cols]
gdf_bikes_end.head()
```

```python
# update to more appropriate data types for each of the columns for clustering

gdf_bikes_end.endPlaceId = gdf_bikes_end.endPlaceId.astype(int)
gdf_bikes_end.endPlaceName = gdf_bikes_end.endPlaceName.astype(str)
gdf_bikes_end['endDate'] = pd.to_datetime(gdf_bikes_end['endDate'], format='%Y-%m-%dT%H:%M:%SZ')

gdf_bikes_end.info()
```

```python
# clustering step: set for 4 clusters using the gdf_bikes_end gdf
# This establishes 4 clusters 

from sklearn.cluster import KMeans
num_clusters = 4

kmeans_collection = KMeans(n_clusters=num_clusters, random_state=42)
gdf_bikes_end['kmeans_cluster'] = kmeans_collection.fit_predict(gdf_bikes_end[['endPlaceLong', 'endPlaceLat']])
```

```python
# plotting the clusters using leafmap

import leafmap

end_map = leafmap.Map(center=(55.860166, -4.257505),
                zoom=12,
                draw_control=False,
                measure_control=False,
                fullscreen_control=False,
                attribution_control=True,
                   
               )

end_map.add_basemap("CartoDB.Positron")
end_map.add_data(
    gdf_bikes_end,
    column='kmeans_cluster',
    legend_title='Clusters',
    cmap='Set2',
    k=4,
)

#Ploting the map
end_map
```

```python
import requests
import pandas as pd
import geopandas as gpd
import matplotlib as plt


# describe the URL
url_sensors = "https://api.glasgow.gov.uk/traffic/v1/movement/sites"
# make query and check that the status of the url is okay 
response = requests.get(url_sensors)
response
```

```python
# translate this into a format that can be manipulated easily
data = response.json()
data

#clear cell output after running to prevent overloading the memory
```

```python
# explore the dataset and checking what columns are present

sensor_pd = pd.DataFrame(data)
sensor_pd.head
```

```python
# collect only the 'from' dictionary and make this a subset

from_data = sensor_pd['from'].apply(pd.Series)
```

```python
# use concat to create the new dataframe with site id and from latitude

sensor_join = pd.concat([sensor_pd['siteId'], from_data], axis=1)

sensor_join.head()
```

```python
# check for null values in the lat and long columns 

nan_in_column_Lat = sensor_join['lat'].isna().any()
nan_in_column_Long = sensor_join['long'].isna().any()

print(nan_in_column_Lat,nan_in_column_Lat)
```

```python
# make into a geodataframe using lat and long columns to make the geometry column 
gdf_sensors = gpd.GeoDataFrame(sensor_join, geometry=gpd.points_from_xy(sensor_join['long'], sensor_join['lat']), crs='EPSG:4326')

gdf_sensors.head()
```

```python
gdf_sensors.explore()
#this shows that there are several outliers in the dataset 
# use the workplace zones layer to crop this to the extent
```

```python
# reading the working sensors data wms

import leafmap 
import requests
import pandas as pd
import geopandas as gpd
import matplotlib as plt

m = leafmap.Map(
    center=(56.329031,-3.798943),
    zoom=7
)
wms_url = 'https://maps.gov.scot/server/services/NRS/Census2011/MapServer/WMSServer?'
# A WMS URL include multiple layers, so you need to provide the name you need to load in your map.
# See this: https://www.spatialdata.gov.scot/geonetwork/srv/eng/catalog.search#/metadata/ff882746-e913-4f78-862e-f6e3974fb80e


m.add_wms_layer(url=wms_url, layers='WorkplaceZones2011', name='Census2011', shown=True)
m
```

as the data is in the format of a wms, this only links to an image or raster file, therefore it cannot be converted to a tabular data format such as a gdf, making any further analysis not possible

```python
working_zones = gpd.read_file('Data/Data_lab2/WorkplaceZones2011Scotland/WorkplaceZones2011Scotland.shp')

#taken from: https://webarchive.nrscotland.gov.uk/20241128123116/https://www.nrscotland.gov.uk/statistics-and-data/geography/our-products/census-datasets/2011-census/2011-boundaries
working_zones.head()
```

```python
glasgow = gpd.read_file("https://services2.arcgis.com/Ne8d9gKn5SJ3eAaw/arcgis/rest/services/Glasgow_City_Council_Boundary/FeatureServer")
```

```python
working_zones['wzID'] = range(0, len(working_zones))
working_zones.head()
```

```python
working_zones.crs
working_zones = working_zones.to_crs('EPSG:4326')
```

```python
import geopandas as gpd

joined = gpd.sjoin(gdf_sensors, working_zones, predicate="intersects", how="left")
```

```python
joined.head()
```

```python
zone_count = joined.groupby('wzID').size().reset_index(name='sensor_count')
zone_count
```

```python
gdf_working_zones = working_zones.merge(zone_count, on='wzID', how='left')
gdf_working_zones.shape
```

```python
glasgow = gdf_working_zones.dropna()
glasgow.shape
```

```python
glasgow.explore()
```
